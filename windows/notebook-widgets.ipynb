```python
import ipywidgets as widgets
from datetime import date

date_picker = widgets.DatePicker(
    description="Target date",
    disabled=False
)

result = widgets.HTML(value="Pick a date")

def update_days(change):
    if change["new"] is None:
        result.value = "<b>Pick a date</b>"
        return
    delta = (change["new"] - date.today()).days
    result.value = f"<b>{delta} day(s)</b>"

date_picker.observe(update_days, names="value")

widgets.VBox([date_picker, result])
```
