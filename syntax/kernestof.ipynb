## Programopbygning

## Parametrisering og abstraktion

Du skal forstå, at parametre kan generalisere en løsning, så samme kode kan løse mange varianter af et problem. Du skal også kunne tænke i abstraktion, hvor du beskriver “hvad” der skal ske, uden at alt låses til én konkret detalje.


```python
def summer(tal, start=0):
    total = start
    for t in tal:
        total += t
    return total

print(summer([1, 2, 3]))
print(summer([1, 2, 3], start=10))
```

Forklaring: Samme funktion kan bruges fleksibelt ved at variere parametre.

## Dataabstraktion

Du skal øve dig i at adskille adfærd fra implementation, så du kan ændre “indmaden” uden at ødelægge resten af programmet. Det hjælper især, når I bygger større programmer eller samarbejder.


```python
class Counter:
    def __init__(self):
        self._value = 0

    def click(self):
        self._value += 1

    def value(self):
        return self._value

t = Counter()
t.click()
print("Værdi: ", t.value())
```

Forklaring: Du bruger en klasse som indpakning og metoder som “grænseflade” og skjuler detaljen (_value) bag klassen.

## Rekursion

Du skal kunne skrive simple rekursive funktioner, genkende problemer der passer til rekursion, og forklare forskellen på rekursion og iteration. 


```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

Forklaring: Funktionen kalder sig selv, indtil den rammer en stop-betingelse.

## Polymorfi 

Bruges ofte fx via nedarvning og dynamisk binding i OOP.

Du skal have en grundforståelse af polymorfi og kunne bruge det til at skrive mere generiske løsninger. I objektorienteret programmering handler det ofte om at kunne behandle forskellige objekter ens, så længe de har samme “adfærd”.


```python
class Dyr:
    def lyd(self):
        raise NotImplementedError

class Hund(Dyr):
    def lyd(self):
        return "vov"

class Kat(Dyr):
    def lyd(self):
        return "mjav"

for d in [Hund(), Kat()]:
    print(d.lyd())
```

Forklaring: Løkken virker for flere typer, fordi de alle tilbyder metoden lyd().

## Algoritmemønstre

Sværhedsgrad: avanceret. 

Du skal kunne genkende og bygge simple mønstre som “tæl”, “find”, “filtrér” og “søg”, og du skal kunne generalisere fra konkrete eksempler. Det hjælper dig med at gå fra “idé” til “kode”, uden at starte forfra hver gang.


```python
def find_første(tal, kriterie):
    for t in tal:
        if kriterie(t):
            return t
    return None

print(find_første([3, 7, 10, 11], lambda x: x % 2 == 0))
```

Forklaring: “Find første der matcher” er et klassisk algoritmemønster, som kan genbruges.

## Overriding 

Begrebet __override__ bruges primært som en teknik i OOP i forbindelse med nedarvning.

En metode kan __overrides__ i en subklasse. Det vil sige, at den adfærd ændres afhængig af instansens type.

Bemærk: Override vs overload. Begrebet __overloading__ - funktion med samme navn men andre parametre - er det ikke samme. Da Python har meget fleksibel håndtering af funktionsparametre, implementeres overloading ofte på anden vis end i andre sprog.


```python
class Base:
    def besked(self):
        return "base"

class Sub(Base):
    def besked(self):   # overriding
        return "sub"

print(Sub().besked())
```

Forklaring: subklassen ændrer adfærden ved at overskrive metoden `besked()`.

## Generiske programdele, biblioteker og API’er

Du skal kunne bruge eksterne biblioteker/API’er og forstå, at de kan gøre komplekse opgaver lettere, og du skal også øve dig i at lave dine egne genbrugelige dele. Et vigtigt aspekt er at kunne dokumentere en API, så andre kan bruge den rigtigt.


```python
import math

def afstand(x1, y1, x2, y2):
    """Returnerer afstanden mellem to punkter i planen."""
    return math.hypot(x2 - x1, y2 - y1)

print(afstand(0, 0, 3, 4))
```

Forklaring: Du bruger et bibliotek (math) og laver samtidig en lille “API” (funktionen) med dokumentation.

## Dokumentation

Du skal kunne beskrive og forklare programmer på et mere abstrakt niveau, og du skal kunne skelne mellem programmets statiske opbygning og dets dynamiske opførsel under kørsel. Du kan bruge fx pseudokode og diagram-tænkning til at planlægge og forklare løsninger, og du skal kunne dokumentere komponenter, så du og andre kan bygge videre.

Konkrete værktøjer er fx

- pseudokode
- flow-diagrammer
- blokdiagrammer


```python
def max_i_liste(tal):
    """
    Pseudokode-idé:
    sæt bedste = første element
    for hvert element: hvis større end bedste, opdater bedste
    returnér bedste
    """
    bedste = tal[0]
    for t in tal[1:]:
        if t > bedste:
            bedste = t
    return bedste

print(max_i_liste([3, 9, 2]))
```

Forklaring: Docstringen viser din “højniveau-plan”, og koden realiserer planen på en struktureret måde.
