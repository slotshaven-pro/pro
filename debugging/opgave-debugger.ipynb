## Opgave 1
Algoritmen i funktionen `fibonacci` beregner fibonacci-værdien for `n`.

Brug debuggeren til at finde ud af hvordan algoritmen fungerer. 

Kald funktionen med værdien 6.

- Sæt et breakpoint i linje 9 (a, b = ...) og aflæs variablern `a` og `b`'s værdier for hver eneste gennemløb 
- Forklar hvad variablerne `a` og `b` repræsenterer (evt. i markdown-celle)
- Når du forstår algoritmen, indsæt kommentarer i koden der forklarer hvordan den virker

- Funktionen indeholder en fejl. Find den ved hjælp af debuggeren. 

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return a

# Calculate F6
n = 6
fn = fibonacci(n)
print(f'The fibonacci number of {n} is {fn}')
```

## Opgave 2 (ekstra opgave, svær)

Algoritmen i funktionen nedenfor finder alle primtal mellem x og y ved hjælp af funktionen "Eratostenes’ si" (eng. _Sieve of Eratosthenes_). 

Brug debuggeren til at forstå algoritmen. 
Forklar i koden med kommentarer hvordan den virker.
Brug notebook eller kopier koden til en python-fil.



```python
# algorithm sieve of eratosthenes
def prime(x, y):
    # insert comment
    primes = [True] * (y + 1)
    # 0 and 1 are not prime
    primes[0], primes[1] = False, False

    # insert comment
    for i in range(2, int(y ** 0.5) + 1):
        if primes[i]:
            # insert comment
            for j in range(i * i, y + 1, i):
                primes[j] = False
    res = [i for i in range(x, y + 1) if primes[i]]
    return res

x, y = 2, 7
res = prime(x, y)
print(res if res else "No")
```
